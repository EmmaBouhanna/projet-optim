### Question 2
En introduisant l'abscisse curviligne d'une courbe et en exploitant le modèle de moteur donné en annexe, écrire la fonction objectif à minimiser

**Réponse:**

* On introduit l'abscisse curviligne d'origine $t_{0}$ comme étant $~
l(t)=\int _{{t_{0}}}^{t}\left\|{v}(\tau )\right\|{\mathrm  d}\tau $



* On note $ T(t) = \tau u(t)$ $\,(1)$ , où $\tau$ est le couple moteur maximal et $u(t)\in [0,1]$ <br>
Le modèle de moteur donné en annexe nous donne les équations suivantes :
  $$\left \{
   \begin{array}{r c l}
      T(t)  & = &  \kappa I_{m}(t) &(2)&\\
      R_{m}I_{m}(t) & = & U_{m}(t) - \kappa \omega(t) &(3)&\\
   \end{array}
   \right .$$
   
   En multipliant chaque membre de l'équation (3) par $I_{m}(t)$, on obtient l'équation suivante :
     $$  R_{m}I_{m}(t)^{2}  =  U_{m}(t)I_{m}(t) - \kappa I_{m}(t) \omega(t)$$
     qu'on peut réécrire :
     $$ \mathcal{P}_{e}(t) = \frac{R_{m}}{\kappa ^{2}}T(t)^{2} + T(t) \omega(t) \quad (4) $$
     où $ \mathcal{P}_{e}(t) = U_{m}(t)I_{m}(t)$ est la puissance d'entrée du moteur


* On veut minimiser l'énergie perdue par le véhicule pendant un tour, c'est-à-dire qu'on veut minimiser $ -\Delta E_{c} $
<br>D'après le théorème de l'énergie cinétique appliqué dans le référentiel lié au sol, $ \Delta E_{c} = \mathcal{P}(\overrightarrow{actions  \, extérieures}) + \mathcal{P}(\overrightarrow{actions \, intérieures})$
<br>Ici, parmi les actions extérieures au système : 
    * le poids est orthogonal au vecteur vitesse et a donc une puissance nulle
    * la puissance de la réaction du sol est nulle d'après l'hypothèse de non-glissemant de la roue
    * la puissance de la force de frottement due à l'air vaut $ - \alpha \int_{t_{0}}^{t_{f}} v(t)^{3} dt $
   

* Le problème d'optimisation qu'on cherche à résoudre est donc le suivant :
$$ \boxed{\underset{l(t_{f}) = L
\\ \mathcal{P}_{e}(t) = \frac{R_{m}}{\kappa ^{2}}T(t)^{2} + T(t) \omega(t)
\\ v_{min} \leq v(t) \leq v_{max}}
{\min \: f(v(t))}}$$
où :
    * la variable de décision est la fonction $\boxed{v: [t_{0}, t_{f}] \rightarrow \mathbb{R}}$ 
    * $\boxed{f(v(t)) =  \int_{t_{0}}^{t_{f}} (\alpha v(t)^{3} - \mathcal{P}_{e}(t)) \: dt}$ 
    * $L$ est la longueur totale de la piste

### Question 3
Sous des hypothèses raisonnables, proposer une relation simple liant vitesse de rotation du rotor du moteur et vitesse du véhicule

* On fait l'hypothèse que **les roues roulent sans glisser sur le sol** et que **les quatre roues ont un comportement identique**. Sur le schéma ci-dessus, $I_{1}$ est un point fixe situé à la périphérie de la roue et $J_{1}$ un point fixe situé sur le sol. Ces deux points sont en contact à l'instant $t$ considéré sur le schéma. On dit qu'il y a non-glissement de la roue sur le sol sur le sol si les vitesses des deux points en contact $I_{1}$ et $J_{1}$ à l'instant $t$ sont égales.
<br>On a donc ici : $\overrightarrow{v}(I_{1}/sol) = \overrightarrow{v}(J_{1}/sol)$
<br> En notant :
    * **$\overrightarrow{\omega_{1}}(t)$ le vecteur vitesse angulaire de rotation de la roue**
    * **$a$ le rayon de la roue**
    
    on a aussi :
    * d'une part, $\overrightarrow{v}(I_{1}/sol) = \overrightarrow{v}(C_{1}/sol)+\overrightarrow{\omega_{1}}(t)\wedge\overrightarrow{C_{1}I_{1}}$
    * d'autre part, $\overrightarrow{v}(J_{1}/sol)$ car $J_{1}$ est un point fixe du sol

    Comme $\overrightarrow{v}(C_{1}/sol)$ peut aussi être notée $\overrightarrow{v}$ la vitesse du véhicule, on a finalement :
$$ v(t) = a \omega_{1}(t) $$


* On fait l'hypothèse que la vitesse de rotation d'une roue est directement proportionnelle à la vitesse de rotation du rotor.
<br>On note **$r$ le rapport de transmission du moteur**, c'est-à-dire $ r = \frac{\omega_{1}(t)}{\omega(t)} $
<br>Dans le cas d'un moteur de voiture, on a $r < 1$

* Ainsi, on obtient une relation simple entre la vitesse de rotation du rotor et la vitesse de déplacement du véhicule :

$$ \boxed{v(t) = a r \omega(t)} $$

### Question 4
Effectuer un bilan des forces agissant sur le véhicule et en déduire l'équation dynamique régissant son mouvement

**Réponse:**

* <u> Système </u> : ensemble de la voiture


* <u> Référentiel </u> : lié au sol, supposé galiléen


* <u> Bilan des forces </u> : 
    * le poids $\overrightarrow{P}$
    * la réaction de contact sur chacune des roues, qu'on peut décomposer comme la somme d'une réaction normale et d'une réaction tangentielle $\overrightarrow{R_{Ci}(t)} = \overrightarrow{R_{Ni}} + \overrightarrow{R_{Ti}(t)}$ (pour chacune des roues $i$)
    * la force de frottement fluide due à l'air $\overrightarrow{F_{air}}=-\alpha \left\|v(t)\right\|\overrightarrow{v(t)} $

Le poids et les réactions normales étant les seules forces normales à la piste, et étant donné qu'on suppose que les roues sont toujours en contact avec la piste, ces forces se compensent.
En revanche, $\overrightarrow{F_{air}}$ et les $\overrightarrow{R_{Ti}(t)}$ sont colinéaires à $\overrightarrow{v(t)}$

On note $\overrightarrow{R_{T}} = R_{T} \frac{\overrightarrow{v(t)}}{\left\|v(t)\right\|}$ la résultante des réactions tangentielles

En appliquant le principe fondamental de la dynamique en utilisant l'abscisse curviligne, on obtient :
$$ m \frac{dv}{dt} = - \alpha v^{2} + R_{T}(t) \quad (5)$$
<br> où $v$ désigne $\left\|v(t)\right\|$

En multipliant chaque membre de l'équation $(5)$ par $v$, on obtient :
$$ \frac{dE_{c}}{dt} = - \alpha v^{3} + R_{T}(t)v(t) $$

En utilisant le théorème de la puissance cinétique et en identifiant, on trouve : $ R_{T}(t)v(t) = \mathcal{P}_{e}(t) $


En substituant dans l'équation $(4)$, on obtient :
$$ R_{T}(t) = \frac{R_{m}}{\kappa ^{2}v(t)}T(t)^{2} + \frac{\omega(t)}{v(t)}T(t) $$

On utilise la relation reliant la vitesse de rotation du rotor à la vitesse du véhicule trouvée à la question 3 :
$$ R_{T}(t) = \frac{R_{m}}{\kappa ^{2}v(t)}T(t)^{2} + \frac{1}{a r}T(t) $$

Enfin, dans l'équation $(5)$, on remplace $R_{T}$ par son expression qu'on vient de trouver, ce qui nous donne l'équation de mouvement du véhicule :
$$ \boxed{ m \frac{dv}{dt} = - \alpha v^{2} + \frac{R_{m}}{\kappa ^{2}v(t)}T(t)^{2} + \frac{1}{a r}T(t)} $$

Le problème d'optimisation devient donc : 
$$ \boxed{\underset{l(t_{f}) = L
\\ v_{min} \leq v(t) \leq v_{max}
\\m \frac{dv}{dt} = - \alpha v^{2} + \frac{R_{m}}{\kappa ^{2}v(t)}T(t)^{2} + \frac{1}{a r}T(t)}
{\min \: f(v(t))}}$$
