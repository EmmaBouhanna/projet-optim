### Question 1
Justifier que, sous des hypothèses raisonnables, on peut réduire le problème au suivant : minimiser l'énergie dépensée pendant un tour de piste (on précisera les contraintes associées à ce nouveau problème)

**Réponse :**

- Hypothèses: 

    - On suppose que la voiture roule sans glisser sur le sol (bitume sans défaut)
    - Les frottements fluides sont quadratiques au vu des valeurs usuelles de vitesse d'une voiture
    - La vitesse est bornée inférieurement et supérieurement
    - On suppose que la puissance électrique initiale est telle qu'on peut effectuer un grand nombre de tour
    
    
- Simplification du problème:

On considère uniquement l'optimisation du profil de vitesse lors d'un tour car lorsque l'on effectue un bilan de puissance, on se rend compte que ce sont les mêmes forces qui interviennent à chaque tour et les mêmes grandeurs qui sont en jeu. Le problème est donc fortement similaire à chaque tour. Ainsi, les conditions sur la vitesse que nous trouverons dans le cadre d'une résolution du problème sur un tour s'appliqueront pour les autres tours de la même façon.

- Contraintes associées à ce nouveau problème :

### Question 2
En introduisant l'abscisse curviligne d'une courbe et en exploitant le modèle de moteur donné en annexe, écrire la fonction objectif à minimiser

**Réponse:**

* On suppose que le circuit est donné sous forme d'arc paramétré, donné par la fonction : $\overrightarrow {OM} =\overrightarrow {g(t)}$, pour $t$ dans l'intervalle $[t_0, t_f]$ où $t_0$ est l'instant où le véhicule commence son tour et $t_f$ celui auquel il le termine.

    On fait l'hypothèse que cet arc paramétré est de classe $\mathcal{C^{1}}$, ce qui est une hypothèse raisonnable étant donné les allures des circuits automobiles.

    On introduit l'abscisse curviligne d'origine $t_{0}$ comme étant
$s(t)=\int _{{t_{0}}}^{t}\left\|{\frac  {{\mathrm  d}\overrightarrow {g}(\tau )}{{\mathrm  d}\tau }}\right\|{\mathrm  d}\tau = \int _{{t_{0}}}^{t}\left\|\overrightarrow {v(\tau)}\right\|{\mathrm  d}\tau$



* On note $ T(t) = \tau u(t)$ $\,(1)$ , où $\tau$ est le couple moteur maximal et $u(t)\in [0,1]$ <br>
Le modèle de moteur donné en annexe nous donne les équations suivantes :
  $$\left \{
   \begin{array}{r c l}
      T(t)  & = &  \kappa I_{m}(t) &(2)&\\
      R_{m}I_{m}(t) & = & U_{m}(t) - \kappa \omega(t) &(3)&\\
   \end{array}
   \right .$$
   
   En multipliant chaque membre de l'équation (3) par $I_{m}(t)$, on obtient l'équation suivante :
     $$  R_{m}I_{m}(t)^{2}  =  U_{m}(t)I_{m}(t) - \kappa I_{m}(t) \omega(t)$$
     On remarque que la puissance électrique fournie par le moteur est en partie perdue par effet Joule ($R_{m}I_{m}(t)^{2}$) et en partie convertie en puissance mécanique ($\kappa I_{m}(t) \omega(t)$).
     
     On peut alors réécrire l'équation précédente uniquement en fonction de $u$ et de $w$:
     $$ \boxed{\mathcal{P}_{e}(t) = \frac{R_{m}}{\kappa ^{2}}(\tau u(t))^{2} + \tau u(t) \omega(t)} \quad (4) $$
     où $ \mathcal{P}_{e}(t) = U_{m}(t)I_{m}(t)$ est la puissance d'entrée du moteur



* On veut minimiser l'énergie perdue par le véhicule pendant un tour, c'est-à-dire qu'on veut minimiser $ -\Delta E_{c} $
    D'après le théorème de l'énergie cinétique appliqué dans le référentiel lié au sol, $ \Delta E_{c} = \int_{t_{0}}^{t_f} (\mathcal{P}(\overrightarrow{actions  \, extérieures}) + \mathcal{P}(\overrightarrow{actions \, intérieures})) \: dt$

    On se place dans le repère mobile d'origine $M(t)$ (la position du véhicule à l'instant $t$) porté par les vecteurs orthonomés $(\overrightarrow{T}, \overrightarrow{N}, \overrightarrow{A})$, où $\overrightarrow{T}$ et $\overrightarrow{N}$ sont respectivement les vecteurs tangentiel unitaire et normal du repère de Frénet lié au véhicule. On considère également le troisième vecteur unitaire du repère cartésien $\overrightarrow{e_z}$, utile par la suite pour projeter le poids sur les vecteurs de la base de Frénet.

    On note $\theta(t)$ l'angle algébrique de pente du circuit à l'abscisse où se trouve le véhicule à l'instant $t$. C'est l'angle entre l'horizontale et la direction de la tangente au circuit (qui est aussi la direction de $\overrightarrow{v}$)

    Le véhicule est modélisé par un point $M(t)$
    
    

* <u> Bilan des forces </u> : 
    * le poids $\overrightarrow{P}$ = $ - mg (\sin{\theta(t)} \overrightarrow{T} + \cos{\theta(t)}\overrightarrow{N}$)
    * la réaction de contact sur chacune des roues, qu'on peut décomposer comme la somme d'une réaction normale à la route et d'une réaction tangentielle $\overrightarrow{R_{Ci}(t)} = R_{Ni}(t) \overrightarrow{N} + R_{Ti}(t) \overrightarrow{T}$ (pour chacune des roues $i$)
    * la force de frottement fluide due à l'air $\overrightarrow{F_{air}}=-\alpha \left\|v(t)\right\|\overrightarrow{v(t)} $ 
    * Dans la suite, on identifie $v(t) = \left\|v(t)\right\|$

* Ici, parmi les actions extérieures au système : 

    * la puissance de la réaction du sol est nulle d'après l'hypothèse de non-glissement de la roue (cf question 3)
    * la puissance du poids à l'instant $t$ vaut $ - mg\sin{\theta(t)}v(t) $
    * la puissance de la force de frottement due à l'air à l'instant $t$ vaut $ - \alpha v(t)^{3} dt $
   

* La fonction de côut du problème d'optimisation qu'on cherche à résoudre est donc la suivante :
$$f(v(t), u(t), w(t)) =  \int_{t_{0}}^{t_{f}} (\alpha v(t)^{3} + mg\sin{\theta(t)}v(t) - \mathcal{P}_{e}(t) ) \: dt$$
que l'on peut réécrire de la façon suivante :
$$\boxed{f(v(t), u(t), w(t)) =  \int_{t_{0}}^{t_{f}} (\alpha v(t)^{3} + mg\sin{\theta(t)}v(t) - \frac{R_{m}}{\kappa ^{2}}(\tau u(t))^{2} - \tau u(t) \omega(t)) \: dt}$$
où :
   * les variables de décision sont pour l'instant les fonctions $\boxed{v: [t_{0}, t_{f}] \rightarrow \mathbb{R}}$, $\boxed{w: [t_{0}, t_{f}] \rightarrow \mathbb{R}}$ et $\boxed{u: [t_{0}, t_{f}] \rightarrow [0,1]}$
   * $L$ est la longueur totale de la piste
   * $t_{0}$ est l'instant où le véhicule démarre son tour de piste
   * $t_{f}$ est l'instant où le véhicule a fait le tour de la piste (il est revenu à son point de départ)
   * la fonction $\theta: [t_{0}, t_{f}] \rightarrow \mathbb{R}$ est supposée connue, car c'est une caractéristique du circuit.
    
Par la suite, on va essayer de trouver des relations reliant $v(t)$, $w(t)$ et $u(t)$ pour se ramener à une seule variable de décision dans le problème continu. Ce dernier sera ainsi piloté par $u$.

### Question 4
Effectuer un bilan des forces agissant sur le véhicule et en déduire l'équation dynamique régissant son mouvement

**Réponse:**

* <u> Système </u> : ensemble de la voiture


* <u> Référentiel </u> : lié au sol, supposé galiléen


* <u> Bilan des forces </u> : 
    * le poids $\overrightarrow{P}$ = $ - mg (\sin{\theta(t)} \overrightarrow{T} + \cos{\theta(t)}\overrightarrow{N}$)
    * la réaction de contact sur chacune des roues, qu'on peut décomposer comme la somme d'une réaction normale à la route et d'une réaction tangentielle $\overrightarrow{R_{Ci}(t)} = R_{Ni}(t) \overrightarrow{N} + R_{Ti}(t) \overrightarrow{T}$ (pour chacune des roues $i$)
    * la force de frottement fluide due à l'air $\overrightarrow{F_{air}}=-\alpha \left\|v(t)\right\|\overrightarrow{v(t)} $ 


$\overrightarrow{F_{air}}$ et les $\overrightarrow{R_{Ti}(t)}$ sont colinéaires à $\overrightarrow{v(t)}$

On note $\overrightarrow{R_{T}} = R_{T} \overrightarrow{T} $ la résultante des réactions tangentielles

En appliquant le principe fondamental de la dynamique et en le projetant sur le vecteur $\overrightarrow{T}$ , on obtient :

$$ \boxed{m \frac{dv}{dt} = - \alpha v(t)^{2} + R_{T}(t) - mg\sin{\theta(t)}} \quad (5)$$
<br> où $v(t)$ désigne $\left\|\overrightarrow{v(t)}\right\|$


En multipliant chaque membre de l'équation $(5)$ par $v$, on obtient :
$$ \frac{dE_{c}}{dt} = - \alpha v(t)^{3} + R_{T}(t)v(t) - mg\sin{\theta(t)}v(t)$$

En utilisant le théorème de la puissance cinétique et en identifiant, on trouve : $ R_{T}(t)v(t) = \mathcal{P}_{e}(t) $


En substituant dans l'équation $(4)$, on obtient :
$$ R_{T}(t) = \frac{R_{m}}{\kappa ^{2}v(t)}T(t)^{2} + \frac{\omega(t)}{v(t)}T(t) $$

On utilise la relation reliant la vitesse de rotation du rotor à la vitesse du véhicule trouvée à la question 3 :
$$ R_{T}(t) = \frac{R_{m}}{\kappa ^{2}v(t)}T(t)^{2} + \frac{1}{a r}T(t) $$

Enfin, dans l'équation $(5)$, on remplace $R_{T}$ par son expression qu'on vient de trouver, ce qui nous donne l'équation de mouvement du véhicule :
$$ m \frac{dv}{dt} = - \alpha v(t)^{2} - mg\sin{\theta(t)} + \frac{R_{m}}{\kappa ^{2}v(t)}T(t)^{2} + \frac{1}{a r}T(t) $$
qu'on peut réécrire en fonction de $u(t)$ :

$$ \boxed{ m \frac{dv}{dt} = - \alpha v(t)^{2} - mg\sin{\theta(t)} + \frac{R_{m}}{\kappa ^{2}v(t)}(\tau u(t))^{2} + \frac{1}{a r}\tau u(t)}\quad (6) $$

On a ainsi établi une relation entre $v(t)$ et $u(t)$ qui étaient nos deux variables de décisions restantes. Maintenant, on choisit d'exprimer toutes nos variables de décision ($v$, $w$, $u$) uniquement en fonction de $u$, car c'est celle qui a le plus de sens physiquement. En effet, il est plus facile de jouer sur $u$ pour optimiser le profil de vitesse de la voiture, que de jouer sur $w$ ou sur $v$ car on a plus de contrôle sur $u$.

### Question 5
On introduit $n$ points de passage équidistants sur le plan de course et on s'intéresse aux temps de passage par ces points. Discrétiser le problème pour le mettre sous la forme
$$\underset{x \in \mathbb{R}^{2n}\\ c(x) \leq 0}{\min \: f(x)}$$
On précisera les variables de décision, les contraintes (au nombre de trois) ainsi que la fonction objectif $f$ à minimiser


**Réponse:**

On note $(t_{1}, ..., t_{n}, t_{n+1})$ les temps d'arrivée respectifs aux $n$ points de passages, avec $t_1$ le point tel que $s(t_1) = 0$, et $t_{n+1}$ le point tel que $s(t_{n+1}) = L$ (le véhicule a fait un tour complet), où $s(t)$ correspond à l'abscisse curviligne de la voiture à l'instant $t$ 
(Si on considère plusieurs tours, le $t_1$ d'un tour correspond au $t_{n+1}$ du tour précédent)
<br>On note $(u_{1}, ..., u_{n})$ les pilotages du couple moteur correspondant à chacun des instants $(t_{1}, ..., t_{n})$
<br>On note $(\theta_{1}, ..., \theta_{n})$ les valeurs des angles de pente du circuit à chacun des points de passage $(1, ..., n)$. On suppose ces valeurs connues ou directement déductibles des équations décrivant le circuit.


On va discrétiser la fonction $f$. Pour cela, on commence par trouver une expression de $f$ en fonction des $u(t_{i})$ et des $v(t_{i})$


- On a d'une part : $$f(v(t),u(t)) =  \sum_{i=1}^{n} (\int_{t_{i}}^{t_{i+1}} (\alpha v(t)^{3} + mg\sin{\theta(t)}v(t) - \mathcal{P}_{e}(t)) \: dt)$$ À l'aide d'un développement limité, on en déduit une expression approchée de $f$ à l'ordre 1 : $$  f(v(t),u(t)) = \sum_{i=1}^{n} (\alpha v(t_{i})^{3} + mg\sin{\theta_i}v(t_i) - \mathcal{P}_{e}(t_{i}) ) \times (t_{i+1} - t_{i})$$


- D'autre part, discrétisons l'équation $(6)$ évaluée en $t_i$ pour $i \in [\![1,n]\!]$ :

    Au premier ordre, par méthode des différences finies, on a : $\frac{dv}{dt} (t_i) = \frac{v(t_{i+1}) - v(t_{i})}{t_{i+1} - t_{i}}$
    On obtient ainsi une relation de récurrence au premier ordre de la vitesse, exprimée en fonction de $u$: 
    
$$ v(t_{i+1}) = v(t_{i}) + [- mg\sin{\theta_i} + \frac{1}{ra} \tau u(t_{i}) - \alpha v(t_{i})^2 + \frac {R_m}{\kappa^2 v(t_{i})} \tau^2 u(t_{i})^2] \times \frac{(t_{i+1} - t_{i})}{m}$$

- Enfin, on reprend la relation $(4)$ et on en déduit :

$$ \mathcal{P}_{e}(t_i) = \frac{R_{m}}{\kappa ^{2}}(\tau u(t_i))^{2} + \tau u(t_i) \times \frac{v(t_i)}{ra} \quad
$$
- On peut ainsi modifier l'expression approchée de f à l'ordre 1 :

$$f(v(t),u(t)) = \sum_{i=1}^{n}(\alpha v(t_{i})^{3} + mg\sin{\theta_i}v(t_{i}) - \frac{R_{m}}{\kappa^{2}}(\tau u(t_{i}))^{2} - \tau u(t_{i}) \frac{v(t_{i})}{ra} ) \times (t_{i+1} - t_{i})$$

On va maintenant essayer d'exprimer les $v(t_i)$ en fonction des $u_i$ et des $t_i$

- On choisit des points <b>équidistants</b>, on doit donc avoir :
$$ \forall i \in [\![1,n+1]\!], \quad s(t_{i})   =   (i-1) \times \frac{L}{n}$$
Cela peut se reformuler de la manière suivante : 
$$ \forall i \in [\![1,n]\!], \quad s(t_{i+1}) - s(t_{i})= \frac{L}{n}$$

- Par différence finie, on peut donner une approximation au premier ordre de $v(t_i)$ à partir de $t_i$, $t_{i+1}$, et les positions à ces temps de passage :

    $$v(t_i)=\frac{s(t_{i+1})-s(t_i)}{t_{i+1}-t_i}$$

   En utilisant le fait que les points de passages sont équidistants, on en déduit :
    
$$ v(t_i)=\frac{L}{n(t_{i+1}-t_i)}$$

   On introduit les variables $\Delta t_{i}$ telles que :
    
$${\Delta t_{i}= t_{i + 1} - t_{i} \quad \forall i \in [\![1,n]\!]}$$

   Ainsi, on peut réécrire les $v(t_i)$:
    
$$v(t_i) = \frac{L}{n\Delta t_{i}}$$


   
On peut ainsi donner une version discrétisée de la fonction f, en écrivant :

   $$f(x) = \sum_{i=1}^{n}(\alpha (\frac{L}{n\Delta t_{i}})^{3} + mg\sin{\theta_i}\frac{L}{n\Delta t_{i}} - \frac{R_{m}}{\kappa^{2}}(\tau u_{i})^{2} - \tau u_{i} \frac{L}{nra\Delta t_{i}})\Delta t_{i} $$
   
qu'on peut simplifier :
$$\boxed{f(x) = \sum_{i=1}^{n}\alpha (\frac{L^{3}}{n^{3}\Delta t_{i}^{2}}) + mg\sin{\theta_i}\frac{L}{n} - \frac{R_{m}}{\kappa^{2}}(\tau u_{i})^{2} - \tau u_{i} \frac{L}{nra}} $$ 
   où $ x = \begin{pmatrix}\Delta t_{1} \\ \vdots \\ \Delta t_{n} \\u_1 \\ \vdots \\u_n\end{pmatrix}$, ce qui nous permet d'avoir pour notre problème de minimisation d'avoir un vecteur et non plus des fonctions comme variable de décision de notre problème de minimisation

Les contraintes du problème d'optimisation peuvent également être réécrites en fonction des $u_i$ et des $s_i$


* La contrainte $v_{min}\leq v(t_i) \leq v_{max} \; \forall i \in [\![1,n]\!]$ peut être réécrite :
$$v_{min}\leq \frac{L}{n\Delta t_{i}} \leq v_{max} \; \forall i \in [\![1,n]\!]$$
ou encore
$$\frac{L}{nv_{max}} \leq \Delta t_{i} \leq \frac{L}{nv_{min}}  \; \forall i \in [\![1,n]\!]$$


* La contrainte $0 \leq u(t_i) \leq 1 \; \forall i \in [\![1,n]\!]$ peut être réécrite :
$$0 \leq u_i\leq 1 \; \forall i \in [\![1,n]\!]$$


* La contrainte liée à l'équation du mouvement $ v(t_{i+1}) = v(t_{i}) + [- mg\sin{\theta_i} + \frac{1}{ra} \tau u(t_{i}) - \alpha v(t_{i})^2 + \frac {R_m}{\kappa^2 v(t_{i})} \tau^2 u(t_{i})^2] \times \frac{(t_{i+1} - t_{i})}{m}$ peut s'écrire:

$$ \frac{L}{n\Delta t_{i+1}} = \frac{L}{n\Delta t_{i}} + [- mg\sin{\theta_i} + \frac{1}{ra} \tau u_{i} - \alpha (\frac{L}{n\Delta t_{i}})^2 + \frac {nR_m\Delta t_{i}}{\kappa^2 L} \tau^2 u_{i}^2] \times \frac{\Delta t_{i}}{m} \quad \forall i \in [\![1,n-1]\!]$$
ou encore:
$$\frac{1}{\Delta t_{i+1}} = (1-\frac{\alpha L}{mn}) \times \frac{1}{\Delta t_{i}} + (\frac{n \tau}{L m R a} u_i - \frac{ng \sin{\theta_i}}{L}) \Delta t_{i} + \frac{n^{2} \tau^{2} R_m}{L^{2} \kappa^{2} m} u_i^{2} \Delta t_{i}^{2}\quad \forall i \in [\![1,n-1]\!]$$

Le problème d'optimisation s'écrit donc : 
$$\underset{ x \in \mathbb{R}^{2n}
\\ \frac{1}{\Delta t_{i+1}} = (1-\frac{\alpha L}{mn}) \times \frac{1}{\Delta t_{i}} + (\frac{n \tau}{L m R a} u_i - \frac{ng \sin{\theta_i}}{L}) \Delta t_{i} + \frac{n^{2} \tau^{2} R_m}{L^{2} \kappa^{2} m} u_i^{2} \Delta t_{i}^{2}\quad \forall i \in [\![1,n-1]\!]
\\ v_{min}\leq \frac{L}{n\Delta t_i} \leq v_{max} \; \forall i \in [\![1,n]\!]
\\  0\leq u_i\leq 1  \; \forall i \in [\![1,n]\!] }
{\min \: f(x)}$$

où $f(x) = \sum_{i=1}^{n}\alpha (\frac{L^{3}}{n^{3}\Delta t_{i}^{2}}) + mg\sin{\theta_i}\frac{L}{n} - \frac{R_{m}}{\kappa^{2}}(\tau u_{i})^{2} - \tau u_{i} \frac{L}{nra}$

On peut le réécrire, plus formellement, de la façon suivante :

$$ \boxed{\underset{ x \in \mathbb{R}^{2n}
\\ \frac{1}{\Delta t_{i+1}} - (1-\frac{\alpha L}{mn}) \times \frac{1}{\Delta t_{i}} - (\frac{n \tau}{L m R a} u_i - \frac{ng \sin{\theta_i}}{L}) \Delta t_{i} - \frac{n^{2} \tau^{2} R_m}{L^{2} \kappa^{2} m} u_i^{2} \Delta t_{i}^{2} \;=\; 0\quad \forall i \in [\![1,n-1]\!]
\\ \frac{L}{nv_{max}} - \Delta t_i \; \leq \; 0  \; \forall i \in [\![1,n]\!]
\\ \Delta t_i - \frac{L}{nv_{min}} \; \leq \; 0  \; \forall i \in [\![1,n]\!]
\\ - u_i \; \leq \; 0  \; \forall i \in [\![1,n]\!]
\\ u_i - 1 \; \leq \; 0  \; \forall i \in [\![1,n]\!]}
{\min \: f(x)} }$$



On remarque que la première contrainte est active.


### Données nécessaires à la résolution


Pour pouvoir résoudre ce problème d'optimisation sous contraintes, nous avons besoin de :

- la masse de la voiture $m$

- les valeurs de $R_m$ et $\kappa$
- le rayon d'une roue $a$

- les données liées au circuit, notamment la longueur totale du circuit $L$ et les valeurs des angles de pente $\theta_i$ à chacun des points de discrétisation.

### Éléments de correction

- On prend $n+1$ pas de temps, $t_0$ correspond au début et $t_n$ correspond au temps où on revient à l'origine. On pose également $\sigma_i = \frac{1}{t_{i+1} \, - \, t_i}$
<br> On en déduit par différences finies que:
    
$v_i(\sigma) = l\sigma_i \tag{1}$

$\dot v_i(\sigma) = l \sigma_i (\sigma_i - \sigma_{i-1})\tag{2}$

où $l=\frac{L}{n}$

- On en déduit par discrétisation de l'équation différentielle issue du principe fondamental de la dynamique et grâce à $(1)$ et $(2)$ que:

$u_i(σ) =\frac{1}{h_1}(l\sigma_i(\sigma_i−\sigma_{i−1})+ h_2 l^2 σ_i^2 +h_0+\gamma(x_i))\tag{3}$
où $\gamma (x)=g sin(\alpha(x))$ et $x_i = i \frac{L}{n}$
- On prend en fait en compte juste la puissance électrique du moteur dans notre fonction de coût, qu'on peut alors discrétiser de la façon suivante :

$J(\sigma) = \frac{1}{2} (\frac{P_0(\sigma)}{\sigma_0} + \frac{P_{n-1}\,(\sigma)}{\sigma_{n-1}}) + \sum_{i=1}^{n-2} \frac{P_i(\sigma)}{\sigma_i} \tag {4}$

où $P_i(\sigma) = b_1 u_i(σ)vi(σ) + b_2 u_i(σ)^2 \tag{5}$

    On sépare la somme en deux et on regarde la convexité de chacun des termes, pour ensuite conclure par linéarité (si $f$ et $g$ positives sont convexes, alors $f+g$ aussi):
    
    
- Tout d'abord, $[1] = \frac{lb_1}{h_1} (\sigma^T A \sigma + b) $ où :

    $ A = \begin{pmatrix} \frac{1}{2} (l+l h_2^2 ) & 0 & \cdots & 0&-\frac{1}{2}l \\ -l & (l+lh_2^2) & 0 & \cdots &0  \\ 0 & -l & (l+lh_2^2) & 0 & \cdots \\ \vdots & 0 &\ddots & \ddots  \\ 0 & \cdots & 0 & - \frac{1}{2}l & \frac{1}{2}(l+lh_2^2) \end{pmatrix}$
    et   $b = (n-1) h_0 + \frac{1}{2} \gamma(0) + \frac{1}{2} \gamma(L-\frac{L}{n}) + \sum_{i=1}^{n-2} \gamma(i\frac{L}{n})$

    On remarque aisément que $[1]$ est convexe car deux fois différentiable de hessienne égale à $A^T + A$ qui est symétrique positive ( = deux fois la partie symétrique de A).


- De plus, on sépare $[2]$ en somme de fonctions positives, convexes par rapport à $\sigma$.
    
    Pour cela, on développe le terme $\frac{u_i^2}{\sigma_i}$ :
    $$ \frac{u_i^2}{\sigma_i} = \frac{1}{h_1^2}[l^2 \sigma_i (\sigma_i - \sigma_{i-1})^2
    + l^4 h_2^2 \sigma_i^3 
    + \frac{1}{\sigma_i} (h_0 + \gamma(x_i))^2
    + 2 l^3 h_2 \sigma_i^2 (\sigma_i - \sigma_{i-1})
    + 2 l^2 h_2 (h_0 + \gamma(x_i)) \sigma_i
    + 2 l (h_0 + \gamma(x_i)) (\sigma_i - \sigma_{i-1}) ]$$
    
    En utilisant cette expression, on écrit $[2]$ comme une somme de 6 termes, multipliée par un réel positif :
    
    $$[2]=\frac{b_2}{h_1^2}([3] + [4] + [5] + [6] + [7] + [8])$$ où :

$[3] = \frac{1}{2} l^2 \sigma_0 (\sigma_0 - \sigma_{n-1})^2 + \frac{1}{2} l^2 \sigma_{n-1} (\sigma_{n-1} - \sigma_{n-2})^2 + \sum_{i=1}^{n-2} l^2 \sigma_i (\sigma_i - \sigma_{i-1})^2 $

On se rend compte que chaque terme à l'intérieur de la somme est le carré d'une fonction convexe $f(\sigma) = (\sigma_i - \sigma_{i-1})$ qui est donc convexe (car la composition d'une fonction convexe par une fonction croissante convexe est convexe, et on se place sur les $\sigma_i$ positifs), multiplié par une fonction linéaire en $\sigma$. Le produit est donc convexe (ce n'est pas vrai en général que le produit de deux fonctions convexes est une fonction convexe, mais ça l'est pour le produit d'une fonction convexe avec une fonction affine positive!)
    
$[4] = h_2^2 l^4 (\frac{1}{2} \sigma_0^3 + \frac{1}{2} \sigma_{n-1}^3 + \sum_{i=1}^{n-2}\sigma_i^3)$ 
<br>Ce terme est convexe comme somme de fonctions convexes par rapport à sigma (sur des $\sigma_i$ positifs), multipliée par un réel positif (le fait que le produit d'une fonction convexe par un réel positif soit convexe sera utilisé à de nombreuses reprises par la suite, et ne sera pas rappelé à chaque fois)
    
    
$[5] = \frac{1}{2}\frac{(h_0 + \gamma(0))^2}{\sigma_0}+ \frac{1}{2}\frac{(h_0 + \gamma(L- \frac{L}{n}))^2}{\sigma_{n-1}}+\sum_{i=1}^{n-2}\frac{(h_0 + \gamma(i\frac{L}{n}))^2}{\sigma_i}$ 
<br> Ce terme est convexe par rapport à $\sigma$ par convexité de la fonction inverse pour les réels positifs (ici, $\sigma_i$ est positif!) et par somme de fonctions convexes.
    
    
$[6] = 2l^3 h_2  [\frac{1}{2} \sigma_0^2(\sigma_0-\sigma_{n-1}) + \frac{1}{2} \sigma_{n-1}^2(\sigma_{n-1} - \sigma_{n-2})+ \sum_{i=1}^{n-2} \sigma_i^2 (\sigma_i - \sigma_{i-1})] $ 
<br>Ce terme est également convexe par rapport à $\sigma$ car chaque terme à l'intérieur de la somme est convexe. 
<br> En effet, $[6] = 2 l^3 h_2 \times \sigma^TD^T \begin{pmatrix} \sigma_0^2 \\ \vdots\\ \sigma_{n-1}^2 \end{pmatrix} $ où $D = \begin{pmatrix} \frac{1}{2} & 0 &\cdots&0&-\frac{1}{2} \\ -1 & 1 &0 & \cdots \\ 0 & \ddots & \ddots & \ddots \\ \vdots \\ \cdots & 0 & -1 & 1 &0 \\ \cdots & 0 &0 & -\frac{1}{2} & \frac{1}{2} \end{pmatrix}$ 
<br>On a donc le produit scalaire d'une fonction convexe (la fonction carré des composantes) par une fonction linéaire $D\sigma$ donc la fonction est convexe.
    
    
$[7] = 2 l^2 h_2 \times [\frac{1}{2}\sigma_0 (h_0+ \gamma(0))+ \frac{1}{2}\sigma_{n-1} (h_0+ \gamma(1-\frac{L}{n}))+\sum_{i=1}^{n-2} \sigma_i (h_0+ \gamma(i\frac{L}{n})] $ 
<br> Ce terme est convexe par rapport à $\sigma$ comme somme de fonctions convexes.
    
    
$[8] = 2l \times [\frac{1}{2}  (\sigma_0-\sigma_{n-1}) (h_0 + \gamma(0))+ \frac{1}{2} (\sigma_{n-1} - \sigma_{n-2}) (h_0 + \gamma(1- \frac{L}{n}))+ \sum_{i=1}^{n-2} (\sigma_i - \sigma_{i-1}) (h_0 + \gamma(i \frac{L}{n}))] = (h_0 + \gamma)^T D^T \sigma$ où $h_0 +\gamma = \begin{pmatrix} \gamma(0) + h_0 \\ \vdots \\ \gamma(1-\frac{L}{n}) + h_0 \end{pmatrix}$ et $D$ définie plus haut.
Ce terme est également convexe (même argument que pour $[6]$)
 

En conclusion, en remplaçant  $$\boxed{ J(\sigma) = \frac{lb_1}{h_1} (\sigma^T A \sigma + b) 
+ \frac{b_2}{h_1^2} [ \frac{1}{2} l^2 \sigma_0 (\sigma_0 - \sigma_{n-1})^2 + \frac{1}{2} l^2 \sigma_{n-1} (\sigma_{n-1} - \sigma_{n-2})^2 + \sum_{i=1}^{n-2} l^2 \sigma_i (\sigma_i - \sigma_{i-1})^2 + h_2^2 l^4 (\frac{1}{2} \sigma_0^3 + \frac{1}{2} \sigma_{n-1}^3 + \sum_{i=1}^{n-2}\sigma_i^3) +  \frac{1}{2}\frac{(h_0 + \gamma(0))^2}{\sigma_0}+ \frac{1}{2}\frac{(h_0 + \gamma(L - \frac{L}{n}))^2}{\sigma_{n-1}}+\sum_{i=1}^{n-2}\frac{(h_0 + \gamma(i\frac{L}{n}))^2}{\sigma_i}  +
2 l^3 h_2 \times \sigma^TD^T \begin{pmatrix} \sigma_0^2 \\ \vdots\\ \sigma_{n-1}^2 \end{pmatrix} + 2h_2l^2 \times (\frac{1}{2}\sigma_0 (h_0+ \gamma(0))+ \frac{1}{2}\sigma_{n-1} (h_0+ \gamma(L-\frac{L}{n}))+\sum_{i=1}^{n-2} \sigma_i (h_0+ \gamma(i\frac{L}{n})) + (h_0 + \gamma)^T D^T \sigma]}$$

On remarque alors que $J$ est bien une somme de fonctions convexes positives, donc notre fonction de coût est convexe!

On remarque enfin que la fonction de coût est même $\alpha$ convexe!!!
En effet, $[1] = \frac{lb_1}{h_1} (\sigma^T A \sigma + b) $ peut se réécrire en 
$[1] = \frac{lb_1}{h_1} (\sigma^T (A-(l^2+h_2l)I_n) \sigma + b + \frac{\alpha}{2} ||\sigma||^2) $ où $\alpha = 2*(l^2+h_2l)$

Ainsi, $J-\frac{\alpha}{2} ||.||^2$ est en encore une fonction convexe, ce qui justifie l'$\alpha$-convexité de J.

- Montrons que les contraintes sont convexes par rapport à $\sigma$.

La première contrainte $ \forall i \in [\![1,n]\!],  v_{min} - v_i(\sigma)$ est linéaire par rapport à $\sigma$ 

La deuxième contrainte $ \forall i \in [\![1,n]\!],  v_i(\sigma) -  v_{max}$ est également linéaire par rapport à $\sigma$

La troisième contrainte $ u_i(\sigma) - 1 = h_1^{-1}(\sigma^T E_i \sigma + h_0 + \gamma(i\frac{L}{n}) ) -1$ où $E = (l+l^2 h_2) e_{ii} -l e_{ii-1}$ avec les $e_{ij}$ correspondant aux vecteurs de base de l'espace des matrices. Pour tout i , la contrainte $[3]_i$ est convexe (même raisonnement que pour le terme $[3]$ de la fonction de coût par exemple) donc la contrainte totale est convexe.

La quatrième contrainte $ -u_i(\sigma)$ est également convexe par le même raisonnement.

Enfin la cinquième contrainte $v_0 - v_{n-1} = l(\sigma_0 - \sigma_{n-1})$ est aussi linéaire par rapport à $\sigma$ (elle peut s'écrire $y^T\sigma$ où $y = \begin{pmatrix} 1 \\ 0 \\\vdots \\ 0 \\ -1 \end{pmatrix} $ ) donc convexe.


On a donc un problème convexe, la fonction de coût étant $\alpha$-convexe et les contraintes convexes !

- Existence et unicité du minimum:

La fonction de coût étant $\alpha$-convexe, le minimimum existe et est unique par théorème du cours. Sous contraintes, l' $\alpha$-convexité garantit l'existence d'un minimiseur mais pas son unicité.

- Méthode de résolution adaptée:

    On a un problème convexe. Pour trouver un minimiseur de $J$ sous contraintes, on a a priori deux possibilités:

   - Une méthode des contraintes actives, si le coût est quadratique et les contraintes affines. Or ici ce n'est pas le cas.
    
   - Un algorithme d'Uzawa ou d'Arrow-Hurwicz, qui repose sur l'approximmation d'un point selle (qui existe par convexité du problème et par qualification des contraintes actives), pour ensuite en déduire un minimiseur de J sous contraintes.
    
On va donc implémenter l'algorithme d'Arrow-Hurwicz pour résoudre le problème.

#### Algorithme de résolution

Développons maintenant un algorithme de résolution, fondé sur celui proposé au TP3. Nous allons implémenter un algorithme d'Arrow-Hurwicz à pas variable en utilisant les conditions de Wolfe, dans un souci d'efficacité. Le but est de trouver un point selle, puis d'en déduire un minimiseur de $J$ sous les contraintes définies plus haut.
- Calcul de $\nabla J$ : 

Pour cela, on utilise la relation $\nabla(fg) = g\nabla f + f\nabla g$

$\nabla J = \sum_{i=1}^{n-2} \frac{\nabla P_i}{\sigma_i} + \nabla{(\frac{1}{\sigma_i})}P_i + 0.5* (\frac{\nabla P_0}{\sigma_0} + \nabla{(\frac{1}{\sigma_0})}P_0 + \frac{\nabla P_{n-1}}{\sigma_{n-1}} + \nabla{(\frac{1}{\sigma_{n-1}})}P_{n-1})$

- Calculons ensuite $\nabla P_i$:

$\nabla P_i = b_1l\sigma_i \nabla u_i + b_1lu_i \nabla \sigma_i + 2b_2u_i \nabla u_i$

Le gradient de $u_i$ est beaucoup plus simple à déterminer, on peut alors en déduire $\nabla J$

```python
# définissons J et grad(J)

def u(sigma, L, n, h_1, h_2, gamma): 
    u = np.zeros(n)
    l = L/n
    for i in range(0,n):
        u[i] = (1/h_1)*(l*sigma[i]*(sigma[i]-sigma[i-1]) + h_2*l*l*sigma[i]**2 + h_0 + gamma[i])
    return u

def L_grad_u(sigma, L, n, h_1, h_2, gamma): #liste des gradients des u_i
    Liste = []
    l = L/n
    for i in range (n):
        grad_ui = np.zeros(n)
        grad_ui[i] = (1/h_1)*((2*l + 2*l*l*h_2)*sigma[i] - l*sigma[i-1])
        grad_ui[i-1] = -(l/h_1)*sigma[i]
        Liste.append(grad_ui)
    return Liste



def J(b_1, b_2, h_1, h_2, h_0, L, n, gamma, sigma): #fonctionne (pas d'erreur en tout cas)
    #sigma est un vecteur colonne de taille n
    l = L/n
    U = u(sigma, L, n, h_1, h_2, gamma)
    P = np.zeros(n)
    sum = 0
    for i in range(0,n): #OK car sigma[-1] = sigma[n-1]
        P[i] = b_1*U[i]*l*sigma[i] + b_2*(U[i])**2
        sum+= P[i]/sigma[i]
    return (-0.5*(1/sigma[0])*P[0] - 0.5*(1/sigma[-1])*P[n-1] + sum)

def P_i(i, sigma, L, n, h_1, h_2, b_2, gamma):
    U = u(sigma, L, n, h_1, h_2, gamma)
    l = L/n
    return(b_1*U[i]*l*sigma[i] + b_2*(U[i])**2)

def gradP_i(i, sigma, L, n, h_1, h_2, b_1, b_2, gamma):
    l = L/n
    U = u(sigma, L, n, h_1, h_2, gamma)
    gradU = L_grad_u(sigma, L, n, h_1, h_2, gamma)
    res = b_1*gradU[i]*l*sigma[i] + b_1*l* U[i]* np.array([0 for _ in range(i)]+[1]+[0 for _ in range(i+1, n)]) + 2*b_2*U[i]*gradU[i]
    return(res)
    
def grad_J(b_1, b_2, h_1, h_2, L, n, sigma, gamma):
    grad = np.zeros(n)
    l = L/n
    U = u(sigma, L, n, h_1, h_2, gamma)
    for i in range(1, n-1):
        grad += gradP_i(i, sigma, L, n, h_1, h_2, b_1, b_2, gamma)/sigma[i]
        grad += P_i(i, sigma, L, n, h_1, h_2, b_2, gamma)*np.array([0 for _ in range(i)] + [-1/sigma[i]**2] + [0 for _ in range(i+1, n)])
    grad += 0.5*gradP_i(0, sigma, L, n, h_1, h_2, b_1, b_2, gamma)/sigma[0]
    grad += 0.5*P_i(0, sigma, L, n, h_1, h_2, b_2, gamma)*np.array([-1/sigma[0]**2]+[0 for _ in range(n-1)])
    grad += 0.5*gradP_i(n-1, sigma, L, n, h_1, h_2, b_1, b_2, gamma)/sigma[-1]
    grad += 0.5*P_i(n-1, sigma, L, n, h_1, h_2, b_2, gamma)*np.array([0 for _ in range(n-1)] + [-1/sigma[-1]**2])
    return(grad)
    
 
```

- On implémente maintenant les contraintes de notre problème, que l'on va concaténer dans un seul vecteur, de façon à respecter les notations du cours.

```python
#définissons les contraintes actives et leur gradient


def contrainte(sigma, L, n, h_1, h_2, gamma):
    l = L/n
    Liste_contraintes = []
    Liste_contraintes.append((sigma[0] - sigma[-1])*(L/n)) #contrainte v0 = v(n-1)
    U = u(sigma, L, n, h_1, h_2, gamma)
    for i in range(n):
        Liste_contraintes.append((U-1)[i]) #contrainte u_i <= 1
    for i in range(n):
        Liste_contraintes.append(-U[i]) #contrainte u_i >= 0
    for i in range(0,n):
        Liste_contraintes.append(v_min - l*sigma[i]) #contrainte v_i >= v_min
    for i in range(0,n):
        Liste_contraintes.append(l*sigma[i] - v_max) #contrainte v_i <= v_max
    return(np.array(Liste_contraintes))

def grad_contrainte(sigma, L, n, h_1, h_2, gamma):
    l = L/n
    Liste_grad = []
    
    Liste_grad.append(np.array([1.0] + [0.0 for _ in range(n-2)] + [-1.0])) #contrainte v0 = v(n-1)
    grad_u = L_grad_u(sigma, L, n, h_1, h_2, gamma)
    for i in range(n):
        Liste_grad.append(grad_u[i]) #contrainte u_i <= 1
    for i in range(n):  
        Liste_grad.append(-grad_u[i]) #contrainte u_i >= 0
    I = np.eye(n)
    for i in range(0,n):
        Liste_grad.append(-l*I[:,i]) #contrainte v_i >= v_min
    for i in range(0,n):
        Liste_grad.append(l*I[:,i]) #contrainte v_i <= v_max
    return(np.array(Liste_grad))
```

##### Algorithme d'Arrow-Hurwitz avec pas de Wolfe

On va maintenant tenter de résoudre le même problème en utilisant l'algorithme d'Arrow-Hurwitz qu'on implémente.

```python
def wolfe_step(fun, grad_fun, xk, pk, c1 = 0.25, c2 = 0.75, M = 1000):
    compteur = 0
    li = 1.0
    lmoins = 0
    lplus = np.inf
    while fun(xk + li*pk)>fun(xk) + c1*li*np.dot(np.transpose(grad_fun(xk)), pk) and np.dot(np.transpose(grad_fun(xk + li*pk)), pk)>c2*np.dot(np.transpose(grad_fun(xk)),pk) and compteur<M:
        if fun(xk + li*pk)>fun(xk) + c1*li*np.dot(np.transpose(grad_fun(xk)), pk):
            lplus = li
            li = 0.5*(lplus+lmoins)
        else:
            lmoins = li
            if lplus == np.inf :
                li = 2*li
            else:
                li = 0.5*(lplus+lmoins)
        compteur+=1
    return li


def arrow_hurwicz(fun, grad_fun, c, grad_c, x0, rho, lambda0, max_iter = 1000, epsilon_grad_L = 0.05):
    n = len(x0)
    compteur = 0
    xk = x0
    lambdak = lambda0
    grad_L = grad_fun(xk) + np.matmul( np.transpose(lambdak), grad_c(xk))
    while compteur<max_iter and np.linalg.norm(grad_L)>epsilon_grad_L:
        grad_L = grad_fun(xk) + np.matmul(lambdak,grad_c(xk))
        pk = - grad_L/np.linalg.norm(grad_L)
        lk = wolfe_step(lambda x : fun(x) + np.matmul(lambdak, c(x)), lambda x: grad_fun(x)+ np.matmul(lambdak,grad_c(x)), xk, pk) # car on veut minimiser le LAGRANGIEN ! 
        xk = xk + lk*pk
        for i in range(0, 4*n+1):
            lambdak [i] = max(0, lambdak[i] + (1/rho)*c(xk)[i])
        compteur += 1
        print(f"itération n° {compteur}")
        print("Temps d execution : %s secondes ---" % (time.time() - start_time))
    return(xk, compteur)
```

##### Algorithme d'Arrow-Hurwitz à pas fixe

```python
def uzawa_fixed_step(fun, grad_fun, c, grad_c, x0, l, rho, lambda0, max_iter = 2000, epsilon_grad_L = 0.02):
    n = len(x0)
    compteur = 0
    xk = x0
    lambdak = lambda0
    grad_L = grad_fun(xk) + np.matmul(np.transpose(lambdak), grad_c(xk))
    while compteur<max_iter and np.linalg.norm(grad_L)>epsilon_grad_L:
        grad_L = grad_fun(xk) + np.matmul(lambdak,grad_c(xk))
        pk = - grad_L/np.linalg.norm(grad_L)
        xk = xk + l*pk
        for i in range(0, 4*n+1):
            lambdak [i] = max(0, lambdak[i] + (1/rho)*c(xk)[i])
        compteur += 1
        print(f"itération n° {compteur}")
        print("Temps d execution : %s secondes ---" % (time.time() - start_time))
    return(xk, compteur)
```
